```python
import os 
import pandas as pd
from pathlib import Path 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
MASTER_DIR = Path(r"D:\Automotive\Race Strategy Decision Support System (RSDSS)\Race-Strategy-Decision-Support-System-RSDSS-\Data\Processed\Master\Master_Dataset.csv")
df = pd.read_csv(MASTER_DIR)
print(f"Dataset loaded successfully.")
print(f"Rows: {len(df):,}")
print(f"Columns: {len(df.columns)}")
```

```python
print("Shape:")
display(df.shape)

print("\nColumns:")
display(df.columns)
```

```python
missing = (
    df.isnull()
      .sum()
      .sort_values(ascending=False)
      .to_frame("Missing Values")
)

display(missing)
```

```python
duplicates = df.duplicated().sum()

print(f"Duplicate rows: {duplicates}")
```

```python
df["ValidLap"].value_counts()
```

```python
df["ValidLap"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    ylabel=""
)

plt.title("Valid vs Invalid Laps")
plt.show()
```

```python
season_counts = df["Year"].value_counts().sort_index()

season_counts
```

```python
season_counts.plot(kind="bar")

plt.title("Rows per Season")
plt.ylabel("Rows")

plt.show()
```

```python
race_counts = df.groupby("RaceName").size()

race_counts.sort_values()
```
